# Sigmoid激活函数

```python
import torch
from utils import show_img

x = torch.linspace(-10,10,steps = 50,requires_grad=True)
v = torch.ones_like(x)

def sigmoid(x):
    y = torch.ones_like(x)/(torch.exp(-x) +1)
    y.backward(v)
    g = x.grad.clone()
    x.grad.zero_()
    return y.detach(),g

def tanh(x):
    y = (torch.exp(x) - torch.exp(-x)) / (torch.exp(x) + torch.exp(-x)) 
    y.backward(v)
    g = x.grad.clone()
    x.grad.zero_()
    return y.detach(),g


y1,g1 = sigmoid(x)
y2,g2 = tanh(x)
show_img(x.detach(),[y1,g1],["Sigmoid","Grad"])
show_img(x.detach(),[y2,g2],["TanH","Grad"])
```

# ReLU激活函数

```python
def relu(x):
    y = torch.maximum(torch.zeros_like(x),x)
    y.backward(v)
    g = x.grad.clone()
    x.grad.zero_()
    return y.detach(),g

def leakyRelu(x):
    y = torch.maximum(x*0.1,x)
    y.backward(v)
    g = x.grad.clone()
    x.grad.zero_()
    return y.detach(),g

y1,g1 = relu(x)
y2,g2 = leakyRelu(x)

show_img(x.detach(),[y1,g1],["ReLU","Grad"])
show_img(x.detach(),[g1,g2],["LeakyReLU","Grad"])
```
